## Add random labels
To simulate the data we'll eventually obtain, randomly assign each row
- A label for their primary sector
- A % of their effort spent in that primary sector
- A label of their secondary sector
- A % of their time they spend in that secondary sector

```python
## Adding Labels
labels = ['A','B','C','D',pd.NA]
df['Primary Sector'] = np.random.choice(labels, size = len(df), p = [0.4,0.3,0.2,0.1,0])
df['Primary Sector Weigting'] = np.random.uniform(size = len(df))
df['Secondary Sector'] = np.random.choice(labels, size = len(df), p = [0.1,0.1,0.1,0.1,0.6])
#df['Secondary Sector Weighting'] = df.apply(lambda x: 1 if df['Secondary Sector'] == 'NA' else 1 - df['Primary Sector Weighting'],axis = 0)
df.head()
```

## Build the graph attributes

Mathematically, a graph is a pair $G=(V,E)$ comprisiing of
- $V$, the set of nodes. In our case we'll have two classes of node (for now): Sectors and Organisations
- $E$, the set of edges which are ordered paris of nodes: $E\subset \{(x,y)|(x,y)\in V^2\}$. 

Here, can take advantage of the edge direction. Sectors can point to Organisations (e.g. the sector 'Land Trust' can point to specific organisations e.g. 'Abbotts Ann Community Land Trust').

We can then use these intial connections to build more edges (e.g. connecting two different land trusts together) to create a more rich graph, potentially aiding visualisaation. 

```python
# Create the lists of nodes
G = nx.DiGraph() # create an empty directed graph
G.add_nodes_from(df['Member Name'])
G.add_nodes_from(df['Primary Sector'])
# add edges from the dataframe
edges = [(row['Member Name'], row['Primary Sector']) for index, row in df.iterrows()]
G.add_edges_from(edges)
```

```python
df['Member Name']
#print(edges)
```

## Visualise

```python
# Draw the graph
nx.draw_spring(G)
```

Let's try make it look nice

```python
val = []
for i in range(len(G.nodes)):
    val.append(i)
nx.set_node_attributes(G, val, 'val')
fig = plt.figure()
nx.draw(G, with_labels=True, node_color=val, font_color="whitesmoke")
fig.set_facecolor('gray')
plt.show()
```

## Make the network interactive

To do this we can use the `pyvis` package. It works nicely with `NetworkX`

```python
# import the needed libraries
from pyvis import network as net

# allow for easy visualisation inside Jupyter Notebooks
graph = net.Network(notebook = True) 

# populate the modes and edges data structures
graph.from_nx(G)

# display
graph.show("example.html")
```
